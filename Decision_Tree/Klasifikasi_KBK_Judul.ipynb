# Preprocessing

## Text Cleaning

```python
import re, string

# Cleaning
def cleaning(text):
  text = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});').sub(' ', text)
  # Case folding
  text = (text.lower()) 
  # Trim text (Menghapus semua spasi dari teks kecuali spasi tunggal di antara kata-kata)
  text = text.strip()
  # Remove punctuations, special characters, and double whitespace
  text = re.compile('<.*?>').sub('', text) 
  text = re.compile('[%s]' % re.escape(string.punctuation)).sub(' ', text)
  text = re.sub('\s+', ' ', text)
  # Number removal
  text = re.sub(r'[0-9]', '', text) 
  text = re.sub(r'[^\w\s]', '', str(text).lower().strip())

  return text
```

```python
df['abstrak'] = df['abstrak'].apply(lambda x: cleaning(x))
df['judul'] = df['judul'].apply(lambda x: cleaning(x))
```

# Pembobotan

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

X_train_vectors_tfidf = tfidf_vectorizer.fit_transform(X_train) 
X_test_vectors_tfidf = tfidf_vectorizer.transform(X_test)

print('Train size: ', X_train_vectors_tfidf.shape)
print('Test size: ', X_test_vectors_tfidf.shape)
```

# Resampling 

## SMOTE

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

X_train_resampled, y_train_resampled = SMOTE(random_state=42).fit_resample(X_train_vectors_tfidf, y_train)
```

```python
y_train_resampled.value_counts()
```

# Training Model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

dtc = DecisionTreeClassifier()
dtc.fit(X_train_resampled, y_train_resampled)

y_pred = dtc.predict(X_test_vectors_tfidf)

print('Accuracy score : ', accuracy_score(y_test, y_pred))
print('Precision score : ', precision_score(y_test, y_pred, average='weighted'))
print('Recall score : ', recall_score(y_test, y_pred, average='weighted'))
print('F1 score : ', f1_score(y_test, y_pred, average='weighted'))
```

## Evaluation

```python
from sklearn.metrics import ConfusionMatrixDisplay
from matplotlib import pyplot as plt

ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap='Blues')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
